```python
import networkx as nx
df = data.head(10) # select the first 10 records

# Create a directed graph
G = nx.DiGraph()

# Add nodes for each unique Product vendor and Product
for vendor in df['Product vendor'].unique():
    G.add_node(vendor, color='blue')
for product in df[' Product'].unique():
    G.add_node(product, color='red')

# Add edges for each Product vendor-Product pair
for _, row in df.iterrows():
    G.add_edge(row['Product vendor'], row[' Product'])

# Plot the network graph
fig, ax = plt.subplots()
pos = nx.spring_layout(G, seed=60)
nx.draw_networkx_nodes(G, pos, ax=ax, node_color=[G.nodes[n]['color'] for n in G.nodes()])
nx.draw_networkx_edges(G, pos, ax=ax)
nx.draw_networkx_labels(G, pos, ax=ax)

# Customize the plot
ax.set_title('Product vendor-Product network graph')
ax.axis('off')

# Show the plot
plt.show()
```

# Summary 
*  This is the Network plot, using for text data only
*  in this plot we identify product of the vendor 
*  Red color indicates product and blue color indicates Product Vendor
